```python
import pandas as pd
import random
import numpy as np

random.seed(42)
np.random.seed(42)

print("Libraries Loaded")
```

```python
cities = ["Bangalore","Hyderabad","Chennai","Mumbai",
          "Pune","Delhi","Jaipur","Lucknow","Patna","Indore"]

cities
```

```python
hubs = []

for i in range(25):
    hubs.append([
        i+1,
        random.choice(cities),
        "India",
        round(random.uniform(12,28),6),
        round(random.uniform(72,88),6),
        random.randint(500,2000),
        pd.Timestamp("2018-01-01") + pd.to_timedelta(random.randint(0,2000),"D")
    ])

hubs_df = pd.DataFrame(hubs,columns=[
    "hub_id","city","state","latitude","longitude","hub_capacity","hub_open_date"
])

hubs_df.head()
```

```python
hubs_df.to_csv("hubs.csv",index=False)
print("Hubs file saved")
```

```python
vehicles = ["Bike","Scooter","EV Bike"]

riders = []

for i in range(200):
    riders.append([
        i+1,
        f"Rider_{i+1}",
        round(random.uniform(0.5,8),1),
        random.choice(vehicles),
        random.randint(0,10),
        pd.Timestamp("2019-01-01") + pd.to_timedelta(random.randint(0,1500),"D"),
        random.randint(1,25)
    ])

riders_df = pd.DataFrame(riders,columns=[
    "rider_id","rider_name","experience_years",
    "vehicle_type","vehicle_age","joining_date","hub_id"
])

riders_df.head()
```

```python
riders_df.to_csv("riders.csv",index=False)
```

```python
customers = []

for i in range(3000):
    customers.append([
        i+1,
        random.choice(cities),
        random.randint(500000,600000),
        round(random.uniform(12,28),6),
        round(random.uniform(72,88),6),
        round(random.uniform(2.5,5.0),1)
    ])

customers_df = pd.DataFrame(customers,columns=[
    "customer_id","city","pincode",
    "latitude","longitude","customer_rating"
])

customers_df.head()
```

```python
customers_df.to_csv("customers.csv",index=False)
```

```python
weather_conditions = ["Clear","Rain","Storm","Fog"]

dates = pd.date_range(start="2024-01-01",periods=365)

weather_data = []

for date in dates:
    for city in cities:
        weather_data.append([
            date,
            city,
            random.choice(weather_conditions),
            random.randint(20,40),
            random.randint(40,95)
        ])

weather_df = pd.DataFrame(weather_data,columns=[
    "date","city","weather_condition","temperature","humidity"
])

weather_df.head()
```

```python
weather_df.to_csv("weather.csv",index=False)
```

```python
delivery_status_options = ["Delivered","Failed","RTO"]
payment_types = ["COD","Prepaid"]

orders = []

for i in range(10000):

    order_time = pd.Timestamp("2024-01-01") + pd.to_timedelta(random.randint(0,365),"D")
    promised_time = order_time + pd.to_timedelta(random.randint(1,3),"D")
    actual_time = promised_time + pd.to_timedelta(random.randint(-1,3),"D")

    orders.append([
        i+1,
        random.randint(1,3000),
        random.randint(1,25),
        random.randint(1,200),
        order_time,
        promised_time,
        actual_time,
        random.choices(delivery_status_options,weights=[70,15,15])[0],
        random.choice(payment_types),
        round(random.uniform(200,5000),2)
    ])

orders_df = pd.DataFrame(orders,columns=[
    "order_id","customer_id","hub_id","rider_id",
    "order_time","promised_delivery_time","actual_delivery_time",
    "delivery_status","payment_type","order_value"
])

orders_df.head()
```

```python
orders_df.to_csv("orders.csv",index=False)
```
